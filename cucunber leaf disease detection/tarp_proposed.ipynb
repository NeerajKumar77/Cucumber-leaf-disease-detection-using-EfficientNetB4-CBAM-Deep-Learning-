```python
# Full pipeline: Transfer learning (EfficientNetB0 pretrained) + augmentation + regularization + fine-tune + Grad-CAM
import os
import math
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import matplotlib.pyplot as plt

# -----------------------------
# CONFIG - change these paths
# -----------------------------
dataset_path = r"/Volumes/YASH/tarp/archive_data/modified Image/Augmented Image/"   # dataset folder containing class subfolders
model_save_path = r"/Volumes/YASH/tarp/best_model.keras"
 
img_height, img_width = 224, 224
batch_size = 32
initial_epochs = 8    # train classifier head
finetune_epochs = 12  # fine-tune later
fine_tune_at = -30    # unfreeze last N layers (use negative indexing to count from end)
l2_reg = 1e-4
seed = 123

# -----------------------------
# quick checks
# -----------------------------
if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"Dataset folder not found: {dataset_path}")

# -----------------------------
# compute class counts & class weights
# -----------------------------
class_names = sorted([d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))])
if len(class_names) == 0:
    raise ValueError("No class subfolders found in dataset_path")

class_counts = {}
total_images = 0
for i, cname in enumerate(class_names):
    cdir = os.path.join(dataset_path, cname)
    # count image files (basic filtering)
    files = [f for f in os.listdir(cdir) if os.path.isfile(os.path.join(cdir, f)) and f.lower().endswith(('.png','.jpg','.jpeg','.bmp','.tiff'))]
    count = len(files)
    class_counts[cname] = count
    total_images += count

num_classes = len(class_names)
print("Classes and counts:", class_counts, "Total:", total_images)

# simple balanced class weights: total/(num_classes * count)
class_weights = {}
for idx, cname in enumerate(class_names):
    cnt = class_counts[cname]
    if cnt == 0:
        class_weights[idx] = 1.0
    else:
        class_weights[idx] = total_images / (num_classes * cnt)
print("Class weights:", class_weights)

# -----------------------------
# dataset loading (fast)
# -----------------------------
train_ds = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="training",
    seed=seed,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    label_mode="int"   # integers for sparse_categorical_crossentropy
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="validation",
    seed=seed,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    label_mode="int"
)

# store class_names from dataset loader (guaranteed same order)
class_names = train_ds.class_names
print("Class names (ordered):", class_names)

# -----------------------------
# augmentation (GPU-friendly) & preprocessing
# -----------------------------
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.15),
    layers.RandomZoom(0.12),
    layers.RandomTranslation(0.08, 0.08),
    layers.RandomContrast(0.12)
], name="data_augmentation")

# normalization layer (0-255 -> 0-1)
normalization_layer = layers.Rescaling(1.0/255)

# apply augmentation + normalization as part of the model pipeline
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y), num_parallel_calls=AUTOTUNE)
train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)
train_ds = train_ds.shuffle(1024).prefetch(AUTOTUNE)

val_ds = val_ds.map(lambda x, y: (normalization_layer(x), y), num_parallel_calls=AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)

# -----------------------------
# Build model: EfficientNetB0 pretrained + regularized head + attention-lite
# -----------------------------
# Attention-lite: squeeze-and-excite style small block (implemented inline)
def se_block(inputs, ratio=8):
    ch = inputs.shape[-1]
    se = layers.GlobalAveragePooling2D()(inputs)
    se = layers.Dense(ch // ratio, activation='relu', kernel_regularizer=regularizers.l2(l2_reg))(se)
    se = layers.Dense(ch, activation='sigmoid', kernel_regularizer=regularizers.l2(l2_reg))(se)
    se = layers.Reshape((1,1,ch))(se)
    return layers.Multiply()([inputs, se])

base_model = EfficientNetB0(include_top=False, weights="imagenet", input_shape=(img_height, img_width, 3))
base_model.trainable = False   # freeze initially

inputs = layers.Input(shape=(img_height, img_width, 3))
x = base_model(inputs, training=False)
x = se_block(x)                                 # attention-lite
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(l2_reg))(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)

model = models.Model(inputs, outputs)

# compile with a reasonable LR for head training
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
# -----------------------------
# Callbacks
# -----------------------------
callbacks = [
    EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, verbose=1),
    ModelCheckpoint(model_save_path, monitor="val_loss", save_best_only=True, verbose=1)
]


# -----------------------------
# Train head (frozen base)
# -----------------------------
history_head = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=initial_epochs,
    callbacks=callbacks,
    class_weight=class_weights
)

# -----------------------------
# Fine-tune: unfreeze only a small portion
# -----------------------------
# Unfreeze last N layers of base_model (fine_tune_at negative index handling)
if isinstance(fine_tune_at, int) and fine_tune_at < 0:
    # count from end
    fine_tune_at_idx = len(base_model.layers) + fine_tune_at
else:
    fine_tune_at_idx = fine_tune_at if isinstance(fine_tune_at, int) else len(base_model.layers) - 50

for i, layer in enumerate(base_model.layers):
    layer.trainable = (i >= fine_tune_at_idx)

# recompile with lower LR
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_finetune = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=finetune_epochs,
    callbacks=callbacks,
    class_weight=class_weights
)

# -----------------------------
# Plot training results
# -----------------------------
def plot_hist(h1, h2=None):
    acc = h1.history.get('accuracy', [])
    val_acc = h1.history.get('val_accuracy', [])
    loss = h1.history.get('loss', [])
    val_loss = h1.history.get('val_loss', [])
    if h2 is not None:
        acc += h2.history.get('accuracy', [])
        val_acc += h2.history.get('val_accuracy', [])
        loss += h2.history.get('loss', [])
        val_loss += h2.history.get('val_loss', [])
    epochs_range = range(1, len(acc)+1)
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(epochs_range, acc, label='train_acc')
    plt.plot(epochs_range, val_acc, label='val_acc')
    plt.legend(); plt.title('Accuracy')
    plt.subplot(1,2,2)
    plt.plot(epochs_range, loss, label='train_loss')
    plt.plot(epochs_range, val_loss, label='val_loss')
    plt.legend(); plt.title('Loss')
    plt.show()

plot_hist(history_head, history_finetune)

# -----------------------------
# Evaluate final model
# -----------------------------
loss, acc = model.evaluate(val_ds)
print(f"Final validation accuracy: {acc*100:.2f}%")

# -----------------------------
# Grad-CAM visualization function
# -----------------------------
def make_gradcam_heatmap(img_array, model, last_conv_layer_name=None, pred_index=None):
    # find a conv layer if not provided
    if last_conv_layer_name is None:
        # search base_model for last conv2d
        conv_layers = [l.name for l in base_model.layers if isinstance(l, tf.keras.layers.Conv2D)]
        if len(conv_layers) == 0:
            raise ValueError("No Conv2D layer found in base model")
        last_conv_layer_name = conv_layers[-1]

    grad_model = tf.keras.models.Model([model.inputs], [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        loss = predictions[:, pred_index]
    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy(), last_conv_layer_name

# Example Grad-CAM usage on a validation image
for images_batch, labels_batch in val_ds.take(1):
    img = images_batch[0:1]  # shape (1,h,w,3)
    heatmap, conv_name = make_gradcam_heatmap(img, model)
    import matplotlib.cm as cm
    import matplotlib.pyplot as plt
    import numpy as np
    # show original
    plt.figure(figsize=(6,3))
    ax = plt.subplot(1,2,1)
    plt.imshow((img[0].numpy()*255).astype('uint8'))
    plt.title("Input")
    plt.axis('off')
    # overlay heatmap
    ax = plt.subplot(1,2,2)
    img_arr = (img[0].numpy()*255).astype('uint8')
    heatmap_resized = tf.image.resize(heatmap[..., tf.newaxis], (img_height, img_width)).numpy().squeeze()
    cmap = cm.get_cmap("jet")
    colored = cmap(heatmap_resized)[:,:,:3]
    overlayed = (0.4 * colored * 255 + 0.6 * img_arr).astype('uint8')
    plt.imshow(overlayed)
    plt.title("Grad-CAM")
    plt.axis('off')
    plt.show()
    break

# -----------------------------
# Done. Best model already saved by ModelCheckpoint
# -----------------------------
print("Best model stored at:", model_save_path)
```
